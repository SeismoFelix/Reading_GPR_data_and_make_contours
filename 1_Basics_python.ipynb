In Python, we constantly create objects of different flavors and we can either figure out their attributes or use methods they are able to execute. Let's create our first object in python.

```python
#My first object in Python
message = 'HOLA MUNDO'
```

```python
#Let's use one of the most popular methods in python "print" for showing the content of the object message
print(message)

#Now let's try this
print("message")
```

*Could you tell the difference?*

Correct, when we use quote symbols we ask to the print method to show explicitily what we are writing between the quote marks. On the other, hand, without the quote marks, we are asking for showing the content of the object named **message**.

## Attributes and methods 2.0

Methods and attributes can be performed for "third" parties. This is, an object could have in-built attributes and methods or external **functions** can act over the object and extract attributes or perform methods.

```python
#Third party attributes extraction
num_characters = len(message)
print(num_characters)
#Here we obtain an integer type of object

#Third party attributes extraction
cut_character = message[0:3]
print(cut_character)
#Here we obtain another character type of object
```

```python
#In built attributes extraction
prefix = 'H'
info = message.startswith(prefix)
print(info)
#Here we obtain a boolean type of object

#In built methods
new_message = message.lower()
print(new_message)
```

**Keep and eye on the latest way (syntax) for calling an object attributes or methods**

Using a dot after a the name of the object is an standard way for extracting information of any object or for performing any method.
